## Research Question

We ask: How do production budget and movie genre influence a film’s success, as measured by box office revenue and IMDb ratings? Specifically, we will examine whether higher budgets and certain genres are associated with greater financial and critical success. The key variables include production budget (numerical), genre (categorical), box office revenue (numerical), and IMDb rating (numerical). We will use statistical analysis and visualizations, such as correlation and group comparisons, to identify relationships between these variables. This is an inferential data analysis task aimed at understanding how these factors contribute to movie success.


## Background and Prior Work

The film industry is strongly influenced by production budgets, which affect everything from casting and visual effects to marketing and distribution. Different movie genres tend to require different levels of funding. For example, large-scale genres such as epic or adventure films often require expensive sets, costumes, and special effects, making them significantly more costly to produce.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) In contrast, genres like drama or romance typically rely more on storytelling and therefore tend to have lower budgets.

Previous research and industry data show that genre plays a major role in determining production cost. Analyses using datasets from sources such as IMDb and Statista indicate that adventure, science fiction, and fantasy films have some of the highest average budgets, often exceeding tens of millions of dollars, while genres such as horror and romance typically remain below $10 million on average.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Other research highlights a different perspective by examining financial outcomes. Studies on movie profitability show that while high-budget films can generate large revenues, they also involve greater financial risk.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) In contrast, lower-budget genres such as horror often produce high returns relative to their cost.

In addition, data science research using large datasets has shown that genre is frequently used as a key variable when predicting movie success and revenue. These findings support the idea that budget differences across genres are consistent and measurable.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

It is also important to consider limitations in this research. Many datasets are biased toward popular or Western films, and movies often belong to multiple genres, making it difficult to isolate the effect of a single genre. These limitations should be considered when interpreting results.

Overall, prior work shows that movie genre is closely tied to production budget, which directly relates to our project.

## Hypothesis


We believe that higher production budgets will be positively associated with higher box office revenue, but will have a weaker or inconsistent relationship with IMDb ratings. We also expect that action, adventure, and fantasy genres will have higher average budgets and revenues compared to genres like drama or romance. This is because high-budget films typically invest more in visual effects and marketing, which drive revenue, while ratings depend more on storytelling and audience perception rather than spending alone.

The ideal dataset would include each film's title, genre, production budget, release year, and country of production, along with additional variables such as runtime and popularity. https://www.kaggle.com/datasets/danielgrijalvas/movies
This dataset contains 6,820 films with variables such as budget, genre, country, release year, and runtime. Budgets are reported in USD, and the dataset is relatively clean. **Kaggle Movies Dataset (Utkarsh)**
https://www.kaggle.com/datasets/utkarshx27/movies-dataset
This dataset includes 4,801 films with information on budget, genre, original language, and popularity. It is also pre-cleaned with minimal missing values. These variables can be used to explore relationships between budget, genre, and success metrics, and provide additional context for analysis.
