## Research Question

We ask: How do production budget and movie genre influence a film's financial and critical success — measured by box office revenue and IMDb ratings — and does this relationship differ between US and non-US productions, and across the pre- and post-streaming eras (before and after 2015)? Specifically, we examine whether the budget-success relationship has weakened as streaming platforms have disrupted theatrical revenue models, and whether Hollywood's historically dominant global distribution network produces a stronger budget-revenue association compared to non-US film industries. The key variables include production budget (numerical), genre (categorical), country of production (categorical), release year (numerical, used to define pre/post-2015 periods), box office revenue (numerical), and IMDb rating (numerical). We will use correlation analysis, group comparisons, and interaction-based statistical models to examine these relationships. This is an inferential data analysis task that builds on and extends prior work by incorporating geographic and temporal dimensions that the existing literature has largely overlooked.


## Background and Prior Work

The film industry is strongly influenced by production budgets, which affect everything from casting and visual effects to marketing and distribution. Different movie genres tend to require different levels of funding. For example, large-scale genres such as action or adventure films often require expensive sets, costumes, and special effects, making them significantly more costly to produce, while genres like drama or romance typically rely more on storytelling and therefore tend to have lower budgets.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Previous research and industry data show that genre plays a major role in determining production cost. Peer-reviewed work by Shahid and Islam (2023) confirmed that drama films consistently carry lower budgets than action or adventure films, and demonstrated that a genre's popularity at the time of release is a strong predictor of box office returns — underscoring how budget and genre interact to shape financial outcomes.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Other research highlights the financial risk dimension: while high-budget films can generate large revenues, they also carry greater financial risk, whereas lower-budget genres such as horror often yield high returns relative to their investment cost.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) A study on IMDb user voting data further found no significant correlation between production budget and average user ratings, suggesting that audience satisfaction is driven by factors beyond spending — such as storytelling quality and emotional resonance — rather than budget alone.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

Despite this body of work, several important gaps remain. First, most existing studies treat the global film industry as a monolith and focus almost exclusively on Hollywood productions. However, international revenues for US studio films have grown from roughly equal to domestic revenue in 2000 to double the domestic level by 2009, and now account for over 70% of Hollywood's total box office take — indicating that the global market has fundamentally reshaped how budget translates to revenue, and that non-US industries operate under very different distribution conditions.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) Cross-national comparisons are therefore critical but largely absent from the literature.

Second, few studies account for the structural disruption caused by the rise of streaming platforms. Research published in *Applied Economics* found that the launch of Netflix streaming services in the US was associated with a 14–17% reduction in theatrical box office revenues, establishing a clear empirical link between streaming availability and declining theatrical performance.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) This suggests that the budget-revenue relationship measured in pre-streaming studies may no longer hold in the same way post-2015, yet most prior analyses do not account for this temporal shift.

Finally, many existing datasets are biased toward popular or high-budget films, and movies often belong to multiple genres, making it difficult to isolate the effect of a single genre. Our project directly addresses all three of these gaps by incorporating country of production as a comparative variable, by splitting our analysis across pre- and post-2015 time periods, and by examining both financial and critical success metrics simultaneously.

## Hypothesis


We hypothesize that the positive relationship between production budget and box office revenue is stronger for US films than for non-US films, because Hollywood benefits from an established global distribution and marketing infrastructure that amplifies returns on high-budget productions — a network advantage that non-US industries generally lack. We further hypothesize that this budget-revenue association has weakened in the post-2015 period compared to the pre-2015 era, as the proliferation of streaming platforms has reduced audiences' dependence on theatrical releases, particularly for lower-budget films that would previously have relied on box office performance as their primary revenue source. In contrast, we expect IMDb ratings to remain largely independent of production budget regardless of country of production or time period, consistent with prior empirical findings showing no significant correlation between budget and average user ratings — suggesting that audience satisfaction is driven more by storytelling quality and emotional resonance than by spending. Among genres, we expect action, adventure, and fantasy films to show the strongest positive budget-revenue correlation, since these genres depend most heavily on spectacle and wide theatrical distribution to justify their investment, while drama and horror — genres that can succeed critically and commercially on low budgets — will show weaker or more variable correlations.


## Data


### Ideal Dataset
The ideal dataset would include each film's title, genre, production budget, release year, and country of production, along with additional variables such as runtime and popularity. A dataset containing approximately 5,000 to 10,000 films across multiple decades would provide sufficient variation for meaningful statistical analysis. These data would be collected from publicly available sources such as IMDb and Kaggle, and combined into a single structured file (e.g., CSV) for analysis. Crucially, the dataset must include the `country` and `release year` fields needed to conduct our US vs. non-US and pre/post-2015 comparisons, as well as both box office revenue and IMDb rating to allow simultaneous analysis of financial and critical success.

### Potential Real Datasets

**Kaggle Dataset (Daniel Grijalvas)**
https://www.kaggle.com/datasets/danielgrijalvas/movies
This dataset contains 6,820 films with variables such as budget, genre, country, release year, gross revenue, and IMDb score. Budgets are reported in USD, and the dataset is relatively clean. The `country` and `year` fields allow us to split the data by US vs. non-US origin and by pre/post-2015 period, directly supporting our comparative research questions.

**Kaggle Movies Dataset (Utkarsh)**
https://www.kaggle.com/datasets/utkarshx27/movies-dataset
This dataset includes 4,801 films with information on budget, genre, original language, and popularity. It is pre-cleaned with minimal missing values. These variables complement the first dataset and allow for additional subgroup analysis and cross-validation of our findings.


## Ethics 

### A. Data Collection
- [X] **A.1 Informed consent**  
This project uses publicly available datasets (e.g., IMDb, Kaggle), so no human subjects are directly involved and informed consent is not required.

- [X] **A.2 Collection bias**  
Movie datasets are biased toward popular, Western, and high-budget films, which may underrepresent non-US industries such as Bollywood, Nollywood, or Korean cinema. This directly affects our US vs. non-US comparison and we will explicitly report sample sizes by country group and acknowledge this imbalance when interpreting results.

- [X] **A.3 Limit PII exposure**  
The datasets used do not include personally identifiable information about individuals. All data points represent films, not people, so privacy risks are minimal.

- [X] **A.4 Downstream bias mitigation**  
Because our datasets skew toward Hollywood productions, conclusions drawn from non-US subgroups may be less statistically reliable. We will flag this limitation clearly in our results and avoid overgeneralizing findings from underrepresented groups. Where possible, we will use proportional sampling or sensitivity analyses to check robustness.

### C. Analysis
- [X] **C.1 Missing perspectives**  
Our datasets significantly underrepresent non-English-language and non-Western film industries. Findings from non-US subgroups should be interpreted cautiously and we will note the limited sample sizes for these groups.

- [X] **C.2 Dataset bias**  
We will examine potential bias in genre representation and budget reporting, including checking whether certain genres or countries are systematically missing or underrepresented in ways that could distort our comparisons.

- [X] **C.3 Honest representation**

- [X] **C.4 Privacy in analysis**

### D. Modeling
- [X] **D.1 Proxy discrimination**  
We will avoid using variables that could serve as proxies for protected characteristics. Country of production is used as a structural variable, not as a proxy for race or ethnicity, and we will be careful not to frame results in ways that imply value judgments about different film industries.

- [X] **D.2 Fairness across groups**  
We will report results separately for US and non-US films and for pre/post-2015 periods rather than aggregating, so that patterns specific to smaller or underrepresented groups are not masked by the dominant Hollywood sample. We will report confidence intervals and note where group sizes are too small for reliable inference.

- [X] **D.3 Metric selection**  
We will carefully interpret metrics such as box office revenue and IMDb ratings and acknowledge their limitations: revenue does not capture streaming income or home video sales, and IMDb ratings may reflect demographic biases in the user base rather than true audience opinion.

- [X] **D.4 Explainability**  
Our methods and results will be clearly explained in plain language alongside technical details. Visualizations will be used to make statistical findings accessible to a non-specialist audience.

- [X] **D.5 Communicate limitations**  
We will explicitly state dataset limitations, potential sources of bias, and boundaries on the generalizability of our conclusions in both our analysis notebooks and final report.
