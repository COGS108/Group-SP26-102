## Research Question

We ask: How do production budget, movie genre, lead actors, language, and country of production influence a film's financial and critical success — measured by box office revenue and IMDb ratings — and does this relationship differ between US and non-US productions, and across the pre- and post-streaming eras (before and after 2015)? Specifically, we examine whether the budget-success relationship has weakened as streaming platforms have disrupted theatrical revenue models, and whether Hollywood's historically dominant global distribution network produces a stronger budget-revenue association compared to non-US film industries. We are also interested in whether lower-budget international films in certain genres can achieve comparable critical success (IMDb ratings) despite lower revenue, suggesting that financial and critical success are not always aligned. The key variables include production budget (numerical), genre (categorical), country of production (categorical), lead actor (categorical), original language (categorical), release year (numerical, used to define pre/post-2015 periods), box office revenue (numerical), and IMDb rating (numerical). We will use correlation analysis, group comparisons, and interaction-based statistical models to examine these relationships. This is an inferential data analysis task that builds on and extends prior work by incorporating geographic, temporal, and talent-related dimensions that the existing literature has largely overlooked.


## Background and Prior Work

The film industry is strongly influenced by production budgets, which affect everything from casting and visual effects to marketing and distribution. Different movie genres tend to require different levels of funding. Research on box office revenue prediction has confirmed that production budget is the single strongest predictor of a film's financial performance, with action and adventure genres consistently commanding the highest budgets while drama and horror operate at far lower cost levels.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Previous peer-reviewed work has also shown that genre plays a major role in determining production cost and financial outcomes. Shahid and Islam (2023) demonstrated that a genre's popularity at the time of release is a strong predictor of box office returns, and that drama films consistently carry lower budgets than action or adventure films — underscoring how budget and genre interact to shape financial outcomes.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

When it comes to critical success, the picture is different. A study analyzing IMDb user voting data found no significant correlation between production budget and average user ratings, suggesting that audience satisfaction is driven by factors such as storytelling quality and emotional resonance rather than spending alone.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) This supports our hypothesis that financial and critical success are not always aligned, and that lower-budget international films may achieve high ratings despite modest revenues.

Despite this body of work, several important gaps remain. First, most existing studies treat the global film industry as a monolith and focus almost exclusively on Hollywood. However, industry data show that international revenues for US studio films now account for over 70% of Hollywood's total box office take, and that non-US industries operate under fundamentally different distribution conditions.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Cross-national comparisons of how budget, language, and talent translate to success are therefore critical but largely absent from the literature.

Second, few studies account for the structural disruption caused by streaming platforms. Research published in *Applied Economics* found that the launch of Netflix streaming in the US was associated with a 14–17% reduction in theatrical box office revenues, establishing a clear empirical link between streaming availability and declining theatrical performance.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) This suggests that the budget-revenue relationship documented in pre-streaming studies may no longer hold post-2015, yet most prior analyses ignore this temporal shift. Our project directly addresses both gaps by incorporating country of production and original language as comparative variables, and by splitting our analysis across pre- and post-2015 time periods.

## Hypothesis


We hypothesize that movies with larger production budgets, globally recognized actors, and English-language releases will generally earn higher box office revenue due to wider international distribution and marketing reach. Specifically, we expect this budget-revenue relationship to be stronger for US films than for non-US films, because Hollywood benefits from an established global distribution and marketing infrastructure that amplifies returns on high-budget productions — a network advantage that non-US industries generally lack.

We further hypothesize that this budget-revenue association has weakened in the post-2015 period compared to the pre-2015 era, as the proliferation of streaming platforms has reduced audiences' dependence on theatrical releases, particularly for lower-budget films that would previously have relied on box office performance as their primary revenue source.

In contrast, we expect IMDb ratings to remain largely independent of production budget regardless of country of production or time period, consistent with prior empirical findings showing no significant correlation between budget and average user ratings. We also predict that some lower-budget international films, particularly in drama or thriller genres, may receive disproportionately high IMDb ratings relative to their revenue — suggesting that critical success and financial success are not always strongly aligned.

Among genres, we expect action, adventure, and fantasy films to show the strongest positive budget-revenue correlation, since these genres depend most heavily on spectacle and wide theatrical distribution to justify their investment, while drama and horror — genres that can succeed critically and commercially on low budgets — will show weaker or more variable correlations.

### Dataset #1 — Movie Industry (Daniel Grijalvas)

This dataset contains 6,820 films released between 1986 and 2016 with 15 variables covering production details, financial performance, and audience reception. The core financial variables are `budget` (production cost in USD) and `gross` (total theatrical box office revenue in USD). The `score` variable is an IMDb user rating on a 0–10 scale and serves as our primary measure of critical/audience success.

Budget values range widely — from a few thousand dollars for micro-budget films to over \$300 million for major blockbusters — with a heavily right-skewed distribution. We address this by log-transforming both budget and gross after adjusting for inflation using BLS CPI-U data (base year: 2020), since raw dollar values are not comparable across four decades. A significant number of entries have missing or zero budget values (likely unreported rather than truly zero-budget). We drop these rows and document the extent of missingness below. Votes are excluded from predictive modeling since they are a post-release variable.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno

# --- Load raw data ---
df = pd.read_csv('data/00-raw/movies.csv')

print('Shape:', df.shape)
df.head()
```

```python
# --- Visualize missingness ---
msno.matrix(df, figsize=(12, 5))
plt.title('Missingness Matrix — Dataset 1 (Raw)')
plt.tight_layout()
plt.show()

missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
print(pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})[missing > 0])
```

```python
# --- Parse release date, drop post-release variable ---
df['release_date'] = pd.to_datetime(
    df['released'].str.extract(r'^([^(]+)').squeeze().str.strip(),
    errors='coerce'
)
df = df.drop(columns=['released', 'votes'])

# --- Strip whitespace from string columns ---
str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda s: s.str.strip())

# --- Cast numeric columns ---
for col in ['budget', 'gross', 'runtime']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# --- Flag zero budgets/gross as missing ---
df.loc[df['budget'] == 0, 'budget'] = np.nan
df.loc[df['gross']  == 0, 'gross']  = np.nan

print('After cleaning, shape:', df.shape)
print('Missing budget:', df['budget'].isnull().sum())
print('Missing gross:', df['gross'].isnull().sum())
```

```python
# --- Drop rows missing budget (our primary predictor) ---
df = df.dropna(subset=['budget'])
print('After dropping missing budget, shape:', df.shape)

# --- CPI-U inflation adjustment to 2020 dollars ---
cpi = {
    1980: 82.4,  1981: 90.9,  1982: 96.5,  1983: 99.6,  1984: 103.9,
    1985: 107.6, 1986: 109.6, 1987: 113.6, 1988: 118.3, 1989: 124.0,
    1990: 130.7, 1991: 136.2, 1992: 140.3, 1993: 144.5, 1994: 148.2,
    1995: 152.4, 1996: 156.9, 1997: 160.5, 1998: 163.0, 1999: 166.6,
    2000: 172.2, 2001: 177.1, 2002: 179.9, 2003: 184.0, 2004: 188.9,
    2005: 195.3, 2006: 201.6, 2007: 207.3, 2008: 215.3, 2009: 214.5,
    2010: 218.1, 2011: 224.9, 2012: 229.6, 2013: 233.0, 2014: 236.7,
    2015: 237.0, 2016: 240.0, 2017: 245.1, 2018: 251.1, 2019: 255.7,
    2020: 258.8
}
BASE_CPI = cpi[2020]

df['cpi']         = df['year'].map(cpi)
df['budget_real'] = df['budget'] * (BASE_CPI / df['cpi'])
df['gross_real']  = df['gross']  * (BASE_CPI / df['cpi'])
df = df.drop(columns=['cpi'])

# --- Log-transform to handle right skew ---
df['log_budget'] = np.log1p(df['budget_real'])
df['log_gross']  = np.log1p(df['gross_real'])

# --- Create era and US indicator columns ---
df['era']     = df['year'].apply(lambda y: 'post_2015' if y >= 2015 else 'pre_2015')
df['us_film'] = df['country'].apply(lambda c: 'US' if c == 'United States' else 'non-US')

print('Final Dataset 1 shape:', df.shape)
df.describe()
```

```python
# --- Visualize distributions to check for outliers ---
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

df['log_budget'].hist(bins=40, ax=axes[0], color='steelblue', edgecolor='white')
axes[0].set_title('Log Budget (2020 USD)')
axes[0].set_xlabel('log(budget)')

df['log_gross'].hist(bins=40, ax=axes[1], color='coral', edgecolor='white')
axes[1].set_title('Log Gross Revenue (2020 USD)')
axes[1].set_xlabel('log(gross)')

df['score'].hist(bins=30, ax=axes[2], color='mediumseagreen', edgecolor='white')
axes[2].set_title('IMDb Score Distribution')
axes[2].set_xlabel('IMDb Score (0-10)')

plt.tight_layout()
plt.show()
```

## Ethics

### A. Data Collection
- [X] **A.1 Informed consent**  
This project uses publicly available datasets (IMDb via Kaggle), so no human subjects are directly involved and informed consent is not required.

- [X] **A.2 Collection bias**  
Both datasets are biased toward popular, Western, and high-budget films, which may underrepresent non-US industries such as Bollywood, Nollywood, or Korean cinema. This directly affects our US vs. non-US comparison. We explicitly report sample sizes by country group and acknowledge this imbalance when interpreting results.

- [X] **A.3 Limit PII exposure**  
The datasets do not include personally identifiable information. All data points represent films, not individuals, so privacy risks are minimal.

- [X] **A.4 Downstream bias mitigation**  
Because our datasets skew toward Hollywood productions, conclusions drawn from non-US subgroups may be less statistically reliable. We flag this limitation in our results and avoid overgeneralizing from underrepresented groups. Where possible, we use proportional sampling or sensitivity analyses to check robustness.
