## Research Question

We ask: How do production budget and movie genre influence a film's success — measured by box office revenue and IMDb ratings — and does this relationship differ between US and non-US films, and across time periods (pre- vs. post-2015)? Specifically, we examine whether the association between budget, genre, and success metrics has shifted with the rise of streaming platforms, and whether these dynamics operate differently across film industries. The key variables include production budget (numerical), genre (categorical), country of production (categorical), release year (numerical), box office revenue (numerical), and IMDb rating (numerical). We will use correlation analysis, group comparisons, and interaction-based statistical models to identify these relationships. This is an inferential data analysis task.


## Background and Prior Work

The film industry is strongly influenced by production budgets, which affect everything from casting and visual effects to marketing and distribution. Different movie genres tend to require different levels of funding. For example, large-scale genres such as epic or adventure films often require expensive sets, costumes, and special effects, making them significantly more costly to produce.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) In contrast, genres like drama or romance typically rely more on storytelling and therefore tend to have lower budgets.

Previous research and industry data show that genre plays a major role in determining production cost. Analyses using datasets from sources such as IMDb and Statista indicate that adventure, science fiction, and fantasy films have some of the highest average budgets, often exceeding tens of millions of dollars, while genres such as horror and romance typically remain below $10 million on average.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Other research highlights a different perspective by examining financial outcomes. Studies on movie profitability show that while high-budget films can generate large revenues, they also involve greater financial risk.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) In contrast, lower-budget genres such as horror often produce high returns relative to their cost.

In addition, data science research using large datasets has shown that genre is frequently used as a key variable when predicting movie success and revenue. These findings support the idea that budget differences across genres are consistent and measurable.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

However, important gaps remain in this literature. First, most existing studies treat the global film industry as a monolith, focusing predominantly on Hollywood productions and rarely comparing US films to non-Western industries such as Bollywood or Korean cinema. This limits the generalizability of their conclusions. Second, few studies account for the structural disruption caused by the rise of streaming platforms around 2015, which fundamentally altered theatrical release strategies and revenue dynamics. A film's box office performance in the pre-streaming era may not be comparable to post-2015 outcomes, yet most prior analyses do not account for this temporal shift. Finally, many datasets are biased toward popular or high-budget films, and movies often belong to multiple genres, making it difficult to isolate the effect of a single genre. Our project addresses these gaps by incorporating country of production as a comparative variable and by examining whether budget-success relationships have shifted across the pre- and post-streaming eras.

## Hypothesis


We hypothesize that the positive relationship between production budget and box office revenue is stronger for US films than for non-US films, because Hollywood benefits from larger global marketing and distribution networks that amplify the returns on high-budget productions. We further hypothesize that this budget-revenue association has weakened in the post-2015 streaming era, as theatrical revenue has become a less reliable indicator of a film's overall financial success. In contrast, we expect IMDb ratings to remain largely independent of budget regardless of country or time period, since audience ratings are more strongly driven by storytelling quality and critical reception than by production spending. Among genres, we expect action, adventure, and fantasy films to show the strongest budget-revenue correlation, while drama and horror genres — which can succeed with lower budgets — will show weaker correlations.


## Data


### Ideal Dataset
The ideal dataset would include each film's title, genre, production budget, release year, and country of production, along with additional variables such as runtime and popularity. A dataset containing approximately 5,000 to 10,000 films across multiple decades would provide sufficient variation for meaningful statistical analysis. These data would be collected from publicly available sources such as IMDb and Kaggle, and combined into a single structured file (e.g., CSV) for analysis.

### Potential Real Datasets

**Kaggle Dataset (Daniel Grijalvas)**
https://www.kaggle.com/datasets/danielgrijalvas/movies
This dataset contains 6,820 films with variables such as budget, genre, country, release year, and runtime. Budgets are reported in USD, and the dataset is relatively clean. The variables 'budget' and 'genre' directly support our research question, while 'country' allows for additional filtering or subgroup analysis.

**Kaggle Movies Dataset (Utkarsh)**
https://www.kaggle.com/datasets/utkarshx27/movies-dataset
This dataset includes 4,801 films with information on budget, genre, original language, and popularity. It is also pre-cleaned with minimal missing values. These variables can be used to explore relationships between budget, genre, and success metrics, and provide additional context for analysis.
